## 1. Variables: Giving Names to Things

A variable is just a name that holds a value. Like a labeled box.

```python
x = 5
```

That's it. We made a box called `x` and put the number `5` in it.

Now we can use `x` and Python knows we mean `5`:

```python
x + 10
```

```python
# We can change what's in the box
x = 100
x
```

```python
# We can have multiple variables
a = 3
b = 4
c = a + b
c
```

```python
# Variables can hold text too (called "strings")
name = "Jason"
name
```

## 4. Functions: Reusable Recipes

A function is like a recipe. You give it ingredients (inputs), it does something, and gives you back a result.

`print()` is a function we've already used!

```python
# Using a built-in function
# len() tells you how long something is

len("hello")   # 5 letters
```

```python
# max() finds the biggest
max(3, 7, 2)   # Returns 7
```

```python
# We can make our own functions!
# Here's a function that doubles a number:

def double(x):
    return x * 2
```

Let's break that down:
- `def` means "I'm defining a function"
- `double` is the name we're giving it
- `(x)` means it takes one input, which we'll call `x`
- `return x * 2` means "give back x times 2"

```python
# Now use it!
double(5)
```

```python
double(100)
```

```python
# A function with two inputs
def add(a, b):
    return a + b

add(3, 4)
```

## 5. Lists: Holding Multiple Things

A list is like a row of boxes. You can put multiple values in order.

```python
numbers = [1, 2, 3, 4, 5]
numbers
```

```python
# Get the first item (counting starts at 0!)
numbers[0]
```

```python
# Get the third item
numbers[2]   # Index 2 = third item (0, 1, 2)
```

```python
# Add something to the end
numbers.append(6)
numbers
```

```python
# How many items?
len(numbers)
```

## 6. Loops: Doing Something Multiple Times

A loop lets you repeat actions.

```python
# Print each number in the list
for n in [1, 2, 3]:
    print(n)
```

What happened:
- First time through: `n` is `1`, we print it
- Second time: `n` is `2`, we print it
- Third time: `n` is `3`, we print it
- No more items, loop ends

```python
# Do math in a loop
for x in [1, 2, 3, 4, 5]:
    print(f"{x} squared is {x ** 2}")
```
